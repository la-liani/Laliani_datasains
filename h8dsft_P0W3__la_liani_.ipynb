  Untuk mengetahui jumlah transaksi dengan status complete setiap bulan selama Q1 hingga Q3 pada tahun 2022 adalah terlebih dahulu mendefenisikan Q1 hingga Q3 adalah dari januari sampai september. untuk transaksi itu sendiri adalah banyaknya order_id yang statusnya complete. untuk menyatakan status complete digunakan kolom delivered_at karena jika menggunakan yang lain seperti created_at itu masih ada kemungkinan untuk NaT. dan perlu untuk diorder by terlebih dahulu. sehingga ditampilkan tiga hasil query dan hasilnya berbeda yang kemungkinan karena ada NaT yang masuk sehingga ikut terhitung sebagai transaksi complete atau karena perbedaan waktu kirim misalnya created_at hari ini tapi delivered_at itu dua hari setelahnya sehinnga itu tidak masuk sebagai range yang ditentukan yakni dari januari sampai september. disini kita menggunakan Subquery, dan insight yang bisa diberikan kita akan lebih cenderung memilih jumlah transaksi berdasarkan kolom delivered_at. hal ini diperkuat dari hasil jumlah transaksinya yang paling kecil yakni 13685. artinya meskipun transaksinya sudah created_at dan delivered_at tapi blum tentu bisa delivered_at.

```python
df2 = client.query("""
SELECT COUNT(order_id) AS jumlah_order
FROM
(SELECT 
   order_id AS order_id,
   created_at AS order_time,
   delivered_at AS send_time,
   shipped_at   AS shipped_time,
       
FROM `bigquery-public-data.thelook_ecommerce.order_items` 

WHERE status = "Complete" 
      AND created_at >= "2022-01-01"
ORDER BY created_at DESC)

WHERE order_time <= "2022-09-30"
      
""").to_dataframe()
df2
```

diatas adalah jumlah transaksi berdasakan kolom created_at

```python
df3 = client.query("""
SELECT COUNT(order_id) AS jumlah_order
FROM
(SELECT 
   order_id AS order_id,
   created_at AS order_time,
   delivered_at AS send_time,
   shipped_at   AS shipped_time,
       
FROM `bigquery-public-data.thelook_ecommerce.order_items` 

WHERE status = "Complete" 
      AND delivered_at >= "2022-01-01"
ORDER BY delivered_at DESC)

WHERE send_time <= "2022-09-30"
      
""").to_dataframe()
df3
```

```python
df2 = client.query(""" # menggunakan subquey untuk dapat menentukan jumlah orderan dengan status complete
SELECT COUNT(order_id) AS jumlah_order
FROM
(SELECT 
   order_id AS order_id,
   created_at AS order_time,
   delivered_at AS send_time,
   shipped_at   AS shipped_time,
       
FROM `bigquery-public-data.thelook_ecommerce.order_items` 

WHERE status = "Complete" 
      AND shipped_at >= "2022-01-01"
ORDER BY shipped_at DESC)

WHERE shipped_time <= "2022-09-30"
      
""").to_dataframe()
df2
```

-Untuk mengetahui total penjualan maka kita akan menghitung total sale_price dengan status complete dari Q1 hingga Q3 selma tahun 2022 berdasarkan kolom delivered_at.

```python
df3 = client.query(""" # sama seperti query sebelumnya tapi kolom yang menjadi perhitungan adalah kolom sale_price dengan status complete
SELECT SUM(total_uang) AS Total_uang
FROM
(SELECT 
   sale_price AS total_uang,
   created_at AS order_time,
   delivered_at AS send_time,
   shipped_at   AS shipped_time,
       
FROM `bigquery-public-data.thelook_ecommerce.order_items` 

WHERE status = "Complete" 
      AND delivered_at >= "2022-01-01"
ORDER BY delivered_at DESC)

WHERE send_time <= "2022-09-30"
      
""").to_dataframe()
df3
```

dari sini kita bisa mengetahui bahwa total hasil transaksi(USD) yang bersifat complete adalah sebesar $799541.180573

- Untuk mengetahui total user yang melakukan transaksi dengan status transaksi yang bersifat complete maka kita akan menghitung jumlah user_id dengan status complete.

```python
df3 = client.query(""" # sama seperti query sebelumnya tapi kolom yang menjadi perhitungan adalah kolom user_id dengan status complete
SELECT COUNT(jumlah_pembeli) AS jumlah_pembeli
FROM
(SELECT 
   user_id AS jumlah_pembeli,
   created_at AS order_time,
   delivered_at AS send_time,
   shipped_at   AS shipped_time,
       
FROM `bigquery-public-data.thelook_ecommerce.order_items` 

WHERE status = "Complete" 
      AND delivered_at >= "2022-01-01"
ORDER BY delivered_at DESC)

WHERE send_time <= "2022-09-30"
      
""").to_dataframe()
df3
```

dapat diketahui jumlah user yang melakukan transaksi dengan status complete sebanyak 13685

menggunakan clausa COUNT DISTINCT untuk cek apakah ada user_id yang berulang.

```python
df3 = client.query(""" # disini kita menggunakan clausa DISTINCT untuk mengetahui apakah ada pembeli yang melakukan pemebelian berulang dan hasilnya ada.
                       #sama seperti query sebelumnya tapi kolom yang menjadi perhitungan adalah kolom user_id dengan status complete
SELECT COUNT(DISTINCT jumlah_pembeli) AS jumlah_pembeli
FROM
(SELECT 
   user_id AS jumlah_pembeli,
   created_at AS order_time,
   delivered_at AS send_time,
   shipped_at   AS shipped_time,
       
FROM `bigquery-public-data.thelook_ecommerce.order_items` 

WHERE status = "Complete" 
      AND delivered_at >= "2022-01-01"
ORDER BY delivered_at DESC)

WHERE send_time <= "2022-09-30"
      
""").to_dataframe()
df3
```

total pembeli yang berbeda adalah sebanyak 8840.
