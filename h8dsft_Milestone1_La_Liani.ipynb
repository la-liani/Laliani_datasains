* Analisis inferensial

```python
salary_mkt_fin = data_5[data_5['specialisation']=='Mkt&Fin']["salary"]
salary_mk_thr = data_5[data_5['specialisation']=='Mkt&HR']["salary"]
```

```python
t_stat, p_val = stats.ttest_ind(salary_mkt_fin,salary_mk_thr) 
print(t_stat)
print(p_val)
```

* uji_t untuk membandingkan rata rata salary specialisasi Mkt&HR dan Mkt&Fin
   * H0_1: μ_salary_Mkt&HR = μ_salary_Mkt&Fin
   * H1_1: μ_salary_Mkt&HR != μ_salary_Mkt&Fin

     berdasarkan hasil uji_t menunjukan gagal tolak H0_1

```python
salary_lk = data_5[data_5['gender']=='F']["salary"]
salary_pr = data_5[data_5['gender']=='M']["salary"]
salary_pr
```

  * uji_t untuk membandingkan rata rata salary laki laki dan perempuan
  * H0_2: μ_salary_lk = μ_salary_pr
  * H1_2: μ_salarylk != μ_salary_pr

```python
t_stat, p_val = stats.ttest_ind(salary_lk,salary_pr) 
print(t_stat)
print(p_val)
```

karena p_val kurang dari 0.05 maka tolak h0_2
